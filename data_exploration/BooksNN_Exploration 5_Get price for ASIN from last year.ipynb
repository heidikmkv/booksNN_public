# Predict price for ASIN from last years' data
### Simplistic extrapolation of price based on last years' prices.

In the end, this script can extract data from the most interesting times of year (determined manually) but this simplistic approach is insufficient.

Note! can get stats from time range in keepa request directly:
https://keepa.com/#!discuss/t/request-products/110

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyisbn #module for converting ISBN-13 to ISBN-10 and vice versa
import booksnn as bnn
```

```python
# GET DATA
asins_in_db = bnn.get_asin_list()
print('Total number of ASINs in database: ' + str(len(asins_in_db)))
asins = asins_in_db[0:10] #select a subset if desired
print('Total number of ASINs selected: ' + str(len(asins))+'\n')

#startdate = '2016-01-01'
#enddate = '2020-05-10'
#df = bnn.dicts_to_dataframe(bnn.get_usedprice(asins),startdate,enddate)

dict_df = bnn.get_usedprice(asins)
```

```python
# Season dates
start_2017 = '2017-01-01'
start_2018 = '2018-01-01'
start_2019 = '2019-01-01'
end_2019 = '2019-12-31'

start_fall_2019 = '2019-08-20'
end_fall_2019 = '2019-09-10'
start_spring_2019 = '2019-01-10'
end_spring_2019 = '2019-02-10'

years = [2017,2018,2019]

#df_summary = pd.DataFrame(columns=['fall','winter','summer'])

for asin,df in dict_df.items():
    #fill in NaNs since the time hasn't been regularized yet
    df = df.interpolate(method='linear')
    df = df.fillna(method='ffill')
    df = df.fillna(method='bfill')
    
    price_data = np.zeros([len(years)+1,5])
    data = {}
    for year in years:
        season_dates = {'sell fall': [str(year) + '-08-15',str(year) + '-09-15'],
                    'buy fall': [str(year) + '-10-01',str(year) + '-12-01'],
                    'sell spr': [str(year) + '-01-01',str(year) + '-02-01'],
                    'buy spr': [str(year) + '-02-10',str(year) + '-05-10']}
                    #,'sell_summer': [str(year) + '-05-01',str(year) + '-05-20']
                    
        prices = []
        for k,v in season_dates.items():
            season_mean = round(df.truncate(v[0],v[1]).mean(),2)['value']
            prices.append(season_mean)
        #add mean and std for year
        #prices.append(round(df[df.index.year==year].mean(),2)['value'])
        #prices.append(round(df[df.index.year==year].std(),2)['value'])
        
        #add % decrease yearly
        from sklearn.linear_model import LinearRegression
        t = df[df.index.year==year].index.to_numpy().astype(np.int64).reshape(-1,1)
        reg_up = LinearRegression().fit(t,df[df.index.year==year]['value'])
        #reg_up.predict(df)
        yearly_change = reg_up.coef_[0]*10**9*60*60*24*365 #reg_coef_ is in units of nanoseconds. Convert to years.
        #print(yearly_change[0])
        prices.append(round(yearly_change,1))
        
        #add recommended buy price
        buy_at = round(season_mean*(100-yearly_change)/100,2)
        prices.append(buy_at)
        
        
        #add next year prediction
        prices.append(round(prices[0]*(100+yearly_change)/100,2))
        
        #add yearly variation
        prices.append(round((prices[4]+1.5*prices[5])/(prices[4]-1.5*prices[5]),2))
        
        data[year] = prices
    df_summary = pd.DataFrame(data).T
    #rename season columns
    i=0
    rename_cols = {}
    for k in season_dates.keys():
        rename_cols[i]=k
        i+=1
    rename_cols[i] = 'yr mean'
    rename_cols[i+1] = 'yr std'
    rename_cols[i+2] = 'linear coeff'
    rename_cols[i+3] = 'yr variation'
    rename_cols[i+3] = 'fall pred'

    df_summary = df_summary.rename(columns=rename_cols)

    # Add seasonality classification
    df_summary = df_summary.reindex(columns=[*df_summary.columns.tolist(), 'fall trend', 'spr trend'], fill_value='-')
    for year in df_summary.index:
        if df_summary.loc[year]['sell fall']/df_summary.loc[year]['buy spr'] > 2:
            df_summary.at[year, 'fall trend'] = 'H'
        elif df_summary.loc[year]['sell fall']/df_summary.loc[year]['buy spr'] > 1.5:
            df_summary.at[year, 'fall trend'] = 'HM'
        
        if df_summary.loc[year]['sell fall']/df_summary.loc[year]['buy spr'] > 2:
            df_summary.at[year, 'spr trend'] = 'H'
        elif df_summary.loc[year]['sell fall']/df_summary.loc[year]['buy spr'] > 1.5:
            df_summary.at[year, 'spr trend'] = 'HM'
        
        if df_summary.loc[year]['fall trend'] == '-' and df_summary.loc[year]['spr trend'] == '-':
            top_prices = df_summary.loc[year]['yr mean'].item()+df_summary.loc[year]['yr std'].item()
            bottom_prices = df_summary.loc[year]['yr mean'].item()-df_summary.loc[year]['yr std'].item()
            df_summary.at[year, 'spr trend'] = 'M'
            df_summary.at[year, 'fall trend'] = 'M'
            
            
    bnn.display_side_by_side(bnn.plot_seasonal(df.truncate(start_2018,end_2019),title=asin),df_summary)
    
```

```python
from scipy.signal import periodogram

f, axs = plt.subplots(len(asins),2,figsize = (18,4*len(asins)))

i=0
for asin,df in dict_df.items():
    bnn.plot_seasonal(df.truncate(start_2017,end_2019),title=asin,ax=axs[i, 0])
    f, Pxx_den = periodogram(np.squeeze(df.resample('D').mean().interpolate(method='linear').to_numpy()),fs=1/365,window='hamming')
    axs[i, 1].semilogy(Pxx_den[1:500])
    #axs[i, 1].xlabel('frequency [Hz]')
    #axs[i, 1].ylabel('PSD [V**2/Hz]')
    #axs[i, 1].show()
    i+=1
    plt.tight_layout()
```

```python
f_spec, ax_spec = plt.subplots()
max_f_index = 200

data_up = np.squeeze(df1.resample('D').mean().interpolate(method='linear').to_numpy())
f_up, S_up = periodogram(data_up,fs=365,window='hamming',scaling='density')
ax_spec.semilogy(f_up[1:max_f_index],S_up[1:max_f_index],color='blue')
ax_spec.set_xlabel('event frequency per year')
#ax_spec.set_ylabel('PSD [V**2/Hz]')

data_sr = np.squeeze(df2.resample('D').mean().interpolate(method='linear').to_numpy())
f_sr, S_sr = periodogram(data_sr,fs=365,window='hamming',scaling='density')
ax_spec.semilogy(f_sr[1:max_f_index],S_sr[1:max_f_index],color='green')

data_uc = np.squeeze(df3.resample('D').mean().interpolate(method='linear').to_numpy())
f_uc, S_uc = periodogram(data_uc,fs=365,window='hamming',scaling='density')
ax_spec.semilogy(f_uc[1:max_f_index],S_uc[1:max_f_index],color='red')

```
